```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import psutil
import os
```

```python
plot_dir = '/research/bsi/projects/PI/tertiary/Hart_Steven_m087494/s211408.DigitalPathology/Quincy/Scripts/CLAM'
```

```python
# Figure 1 - compare test accuracy

# creating the dataset 
acc_compare_dic = {'M1':89, 'M2':84, 'M3':85, 'M4':88,
                   'M5':78, 'M6':89, 'M7':87, 'M8':89,
                   'M9':87, 'M10':88, 'M11':89, 'M12':88,
                   'M13':89, 'M14':88, 'M15':90}


model_names = sorted(acc_compare_dic, key=acc_compare_dic.get, reverse=True)
acc_values = list()
for i in model_names:
    acc_values.append(acc_compare_dic[i])
#print(acc_values)

fig1 = plt.figure(figsize = (8, 8)) 
  
# creating the bar plot 
rects = plt.bar(model_names, acc_values,  
        width = 0.4) 
 
plt.xlabel("Model Names") 
plt.ylabel("Patch-Level Validation Accuracy (%)") 
plt.title("Classification Performance \n Comparison Test on BACH") 

ax = plt.gca()
ax.tick_params(axis='x', colors='blue')
ax.tick_params(axis='y', colors='red')

my_colors = 'rgbkymc'

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., 1.01*height,
                '%d' % int(height),
                ha='center', va='bottom')

autolabel(rects)

plt.tight_layout()

# save plot
plt.savefig(os.path.join(plot_dir, 'BACH_Compare_Test_Figure1.png'))

# visualize plot 
plt.show()
```
