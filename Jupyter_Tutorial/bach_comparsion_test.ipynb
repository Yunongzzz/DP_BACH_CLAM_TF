```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import psutil
import os
```

```python
plot_dir = '/path/'
```

```python
# Figure 1 - compare test accuracy

# creating the dataset 
acc_compare_dic = {'CLAM':87.5000, 'ResNet152':98.4375} 
model_names = list(acc_compare_dic.keys()) 
acc_values = list(acc_compare_dic.values()) 
   
fig1 = plt.figure(figsize = (8, 8)) 
  
# creating the bar plot 
rects = plt.bar(model_names, acc_values, color =['red','blue'],  
        width = 0.4) 
 
plt.xlabel("Model Names") 
plt.ylabel("Test Results in Percentage (%)") 
plt.title("Classification Performance \n Comparison Test on BACH \n CLAM V.S. ResNet152") 

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., 1.01*height,
                '%d' % int(height),
                ha='center', va='bottom')

autolabel(rects)

plt.tight_layout()

# save plot
plt.savefig(os.path.join(plot_dir, 'BACH_Compare_Test_Figure1.png'))

# visualize plot 
plt.show()
```

```python
## Figure 2 - comparison test

# set width of bar 
barWidth = 0.25

fig = plt.subplots(figsize = (9,8)) 

model_names = ['CLAM', 'ResNet152']
metrics_names = ['Accuracy', 'Sensitivity', 'Average of Sensitivity and Accuracy']
# set height of bar 
acc_cr= [87.50000, 98.43750]     # in the order clam, resnet152
sensitivity_cr = [87.50000, 96.87500]
ave_sen_acc_cr = [87.50000, 97.65625]

# Set position of bar on X axis 
br1 = np.arange(len(model_names)) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2]

# Make the plot 
rects1 = plt.bar(br1, acc_cr, color ='r', width = barWidth, 
        edgecolor ='grey', label = metrics_names[0]) 
rects2 = plt.bar(br2, sensitivity_cr, color ='g', width = barWidth, 
        edgecolor ='grey', label = metrics_names[1]) 
rects3 = plt.bar(br3, ave_sen_acc_cr, color ='b', width = barWidth, 
        edgecolor ='grey', label = metrics_names[2]) 
   
# Adding Xticks  
plt.xlabel('Model Names', fontweight ='bold') 
plt.ylabel('Test Results in Percentage (%)', fontweight ='bold') 
plt.xticks([i + barWidth for i in range(len(model_names))], 
           model_names) 
plt.legend(metrics_names)

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., 1.01*height,
                '%d' % int(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.tight_layout()

plt.savefig(os.path.join(plot_dir, 'BACH_Compare_Test_Figure2.png'))

plt.show() 
```
